```python
import pandas as pd

sheet_id = "1bclX5PbqE8zR1-wOSM3ZTFa53sIC7RXqwty7Abmw9yQ"
sheet_name = "2017_jun_funal"  

csv_url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"

#1
df = pd.read_csv(csv_url)

#2
print(f'task 2: \n {df.head()}')

#3
print(f'task 3: \n {df.shape}')

#4
print(f'task 4: \n {df.dtypes}')

#5
missing_values = df.isnull().sum() / len(df) * 100
print(f"task 5: \n {missing_values}")

#6
programming_language_col = df['Мова програмування']
print(f'task 6: \n {programming_language_col}')

#7
missing_values_after_drop = programming_language_col.isnull().sum() / len(df) * 100
print(f"task 7: \n {missing_values_after_drop}")

#8
programming_language_col.dropna(inplace=True)

#9
print(f"task 9: \n {programming_language_col.shape}")

#10
python_data = df[df['Мова програмування'] == 'Python']
print(f'task 10: \n {python_data}')

#11
print(f"task 11: \n {python_data.shape}")

# Завдання 12
grouped_data = df.groupby('Посада')

# Завдання 13
aggregated_data = grouped_data.agg(min_salary=('Зарплата на місяць', 'min'),
                                   max_salary=('Зарплата на місяць', 'max'),
                                   avg_salary=('Зарплата на місяць', 'mean'))
print("\nAggregated data by position:")
print(aggregated_data)

# Завдання 14
# Додавання нового стовпця із середньою зарплатою
df['avg'] = df['Зарплата на місяць']

# Завдання 15
print("\nDescriptive statistics for the 'avg' column:")
print(df['avg'].describe())

# Завдання 16
df.to_csv('output.csv', index=False)
```
